```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, accuracy_score
from sklearn.model_selection import train_test_split
from tensorflow.keras.datasets import mnist
(train_images, train_labels), (test_images, test_labels) = mnist.load_data()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, accuracy_score
import numpy as np

# Flatten the images: each 28x28 image becomes a vector of 784 features
X_train = train_images.reshape(train_images.shape[0], -1)
X_test = test_images.reshape(test_images.shape[0], -1)

# Normalize pixel values to [0, 1]
X_train = X_train / 255.0
X_test = X_test / 255.0

# Labels remain unchanged
y_train = train_labels
y_test = test_labels


```

```python
# Train logistic regression on MNIST
model = LogisticRegression(max_iter=1000, solver='lbfgs', verbose=1)
model.fit(X_train, y_train)
```

```python
# Predict on the test set
y_pred = model.predict(X_test)

# Print accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy * 100:.2f}%")

# Classification report
print(classification_report(y_test, y_pred))
```

```python
# Choose a test instance to visualize
instance = 10
image = test_images[instance]
true_label = test_labels[instance]
predicted_label = y_pred[instance]

# Plot the image with predicted and true labels
plt.imshow(image, cmap='gray')
plt.title(f"True Label: {true_label}, Predicted: {predicted_label}")
plt.axis('off')
plt.show()
```
